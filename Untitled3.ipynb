```python
import pandas
import numpy
from torchvision import datasets, transforms
from torch.utils.data import random_split
from torch.utils.data import DataLoader

train_path = r'C:\Users\kyler\Data Science\Machine Learning\Computer Vision\real_dataset'

train_transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
])

# The data is transformed here, resized into 32 x 32 pixels, flipped and rotated randomly so the model sees the data in different forms and has to generalize
# It's then converted to a tensor so the network can actually accept it

full_data = datasets.ImageFolder(train_path, transform = transform)

train_size = int(0.7*len(full_data))
test_size = len(full_data) - (train_size)

test_data, train_data = random_split(full_data, [train_size, test_size])

train_loader = DataLoader(train_data, batch_size = 32, shuffle = True)
test_loader = DataLoader(test_data, batch_size = 32, shuffle = False)

full_data.classes

# Test and training datasets are created from the real_images folder
# This section loads the train and test loaders and establishes the model architecture. 2 hidden layers + a flatten layer are used. The flatten layer ensures the data is in a single dimension, each pixel side by side
# Referred to classes here to get an idea of the different classes
```

```python
import torch
import torch.nn as nn
 
class SeparableConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1, padding=1):
        super(SeparableConv2d, self).__init__()
        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size=kernel_size,
                                   stride=stride, padding=padding, groups=in_channels)
        self.pointwise = nn.Conv2d(in_channels, out_channels, kernel_size=1)
 
    def forward(self, x):
        x = self.depthwise(x)
        x = self.pointwise(x)
        return x

# Here is the first step of the actual architecture. The separable convolutions are constructed and named SeparableConv2d
# The function is defined, and then the depthwise and pointwise convolutions are defined below
# The forward function tells the model to feed the data X through the model first through the depthwise layer, then pointwise
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = SeparableConv2d(3, 16, 3)
        self.pool = nn.MaxPool2d(2)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(16 * 16 * 16, 512)
        self.relu1 = nn.ReLU()
        self.dropout = nn.Dropout(0.45)
        self.fc2 = nn.Linear(512, 256)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(256, 5)

# Here is the meat of the model architecture.
# Model is defined, and the model is initiated and referred to as self
# Super() is used to pull in functions from the parent function
# SeparableConv2d is called and placed at the start of the network, with 3 input channels (Red, green and blue), 16 output channels, and 3 kernels
# A pooling layer compresses the size of the dimensions of the image, making it less computationally expensive and making the model less sensitive to feature location
# A flatten layer is used to lay the pixels out side by side, making it 1D
# The first layer receives an input of 16 images (equal to the 16 output channels in the SeparableConv2 layer), with each having a 16 x 16 dimension
# 512 output channels are passed to the activation layer
# A dropout layer randomly turns off 45% of the neurons before the next layer
# The next layer receives 512 inputs and produces 256 outputs
# These 256 outputs are put through activation again to ensure non-linearity
# The last layer receives 256 inputs and produces 5 class outputs for the 5 classes of image
# Softmax specification isn't needed because it's built in to CategoricalCrossEntropy

    def forward(self, x):
        x = self.conv1(x)
        x = self.pool(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        return x

# Each layer is called 1 by 1 in the order listed above. 
# The model is told to pass the data (X) through each layer as laid out above


model = Model()
criterion = nn.CrossEntropyLoss() 
optimizer = optim.Adam(model.parameters(), lr=0.005)

# It took a bit of messing around to find a good learning rate for this model. 0.005 seemed to do fine, and the accuracy hovered around 60-70% with different dropout values.

loss_values = []
accuracy_values = []

epochs = 50
for epoch in range(epochs):
    correct = 0
    total = 0
    for images, labels in train_loader:
        predictions = model(images)
        loss = criterion(predictions, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        _, predicted = torch.max(predictions, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    accuracy = correct/total * 100
    loss_values.append(loss.item())
    accuracy_values.append(accuracy)
    print(f"Epoch {epoch+1} - Loss: {loss.item():.4f} - Accuracy: {accuracy:.2f}%")

# This section creates 2 lists, loss_values and accuracy_values, for later plotting.
# A for loop is created. For each epoch in the epochs range, start at a count of 0 correct and total. Then, within each epoch, loop through each pair of images and labels in the previously established train_loader.
# Our loss is calculated using our crossentropy established above, comparing the value of our newly created predictions vs. our established labels.
# We use zero_grad to zero out our calculations after every epoch, to avoid previous epochs' data from interfering with new data.
# We then call loss backward and optimizer.step to backpropagate, our loss function calculating the changes our optimizer needs to make and storing it in the .grad function, so Adam can go in and adjust parameters to nudge our loss lower.
# Below that, we create our calculations for plotting. We pull our predictions out using torch.max to identify what the model thinks is the right label, based on the highest probability value.
# We add those values to our correct and total values, and calculate accuracy below. We append loss and accuracy to our lists, print and plot them.

plt.plot(range(1, epochs + 1), loss_values, marker='o', label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.show()

plt.plot(range(1, epochs + 1), accuracy_values, marker='o', label='Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.show()

correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        predictions = model(images)
        _, predicted = torch.max(predictions, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)
    sample_images, sample_labels = next(iter(train_loader))
    sample_preds = model(sample_images)
    print(sample_preds[:3])
    print(sample_labels[:3])

# Similar process as above, but this allows us to view our total test accuracy, as well as seeing our predictions as we move through each epoch.

print(f"Test Accuracy: {correct / total * 100:.2f}%")

all_predictions = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        predictions = model(images)
        _, predicted = torch.max(predictions, 1)
        all_predictions.extend(predicted.numpy())
        all_labels.extend(labels.numpy())

matrix = confusion_matrix(all_labels, all_predictions)
display = ConfusionMatrixDisplay(matrix, display_labels = full_data.classes)
display.plot()
plt.show()

# Here is where we create our confusion matrix, by creating all_predictions and all_labels lists.
# These lists are appended in a similar way as above, with a for loop and extending current lists.
# confusion_matrix and ConfusionMatrixDisplay were imported from sklearn, and visualization can be seen below.
```
